<h3>Exercise 7: Data Preprocessing Pipeline</h3>


```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import RandomForestClassifier

X = data.drop('music_genre', axis=1)  
y = data['music_genre']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

numeric_features = X.select_dtypes(include=['int64', 'float64']).columns
categorical_features = X.select_dtypes(include=['object']).columns

numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),  
    ('scaler', StandardScaler())  
])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')), 
    ('onehot', OneHotEncoder(handle_unknown='ignore'))  
])

preprocessor = ColumnTransformer(transformers=[
    ('num', numeric_transformer, numeric_features),
    ('cat', categorical_transformer, categorical_features)
])

model_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', RandomForestClassifier())
])

model_pipeline.fit(X_train, y_train)

y_pred = model_pipeline.predict(X_test)
```

<h4>What are the benefits of using a preprocessing pipeline?</h4>
Automation: Pipelines automate the entire preprocessing and modeling process, reducing the chances of errors.
Consistency: Ensures that the same transformations are applied to both the training and testing datasets, leading to a fair and consistent evaluation.
Modularity: Pipelines make it easier to maintain and modify code, allowing you to add or remove steps as needed.
Efficiency: Combining preprocessing and model training into a single workflow reduces the need for repetitive coding.


<h4>How does the pipeline ensure consistency between training and test data transformations?</h4>
The pipeline is first fitted on the training data, learning parameters (e.g., scaling factors, imputed values) from the training set. These same parameters are then applied when transforming the test data, ensuring consistent preprocessing across both datasets.


<h4>How can you extend the pipeline to include additional preprocessing steps?</h4>
You can easily add new steps to the pipeline or ColumnTransformer. For example:
Dimensionality reduction: You can add a PCA or feature selection step before the classifier.
Feature engineering: Add custom feature creation or transformation steps within the pipeline.
Advanced imputation: Add more sophisticated missing value handling techniques, such as KNN imputation.
